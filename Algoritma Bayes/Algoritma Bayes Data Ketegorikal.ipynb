```
def bayesian_learning(df,tuple_X,kolom_attribute,kolom_target):
  def peluang(sample,banyak_data):
    return sample/banyak_data

  #menentukan peluang target
  target = df.groupby(kolom_target)[df.columns[1]].count()
  peluang_target = {}
  for key, value in target.items():
    peluang_target[key] = peluang(value,len(df))
  
  #menentukan prob bersyarat dari tuple
  try:
    prob = {}
    for i in kolom_attribute:
      temp = {}
      att = df.groupby([kolom_target,i])[df.columns[1]].count()
      for key, value in target.items():
        temp[key] = peluang(att[key][tuple_X[i]],value)
      prob[i] = temp
  except Exception:
    prob.clear()
    target_baru = df.groupby(kolom_target)[df.columns[1]].count() + 1
    peluang_target = {}
    #Menghitung ulang peluang target
    for key, value in target_baru.items():
      peluang_target[key] = peluang(value,len(df)+2)
    for i in kolom_attribute:
      temp = {}
      att = df.groupby([kolom_target,i])[df.columns[1]].count() + 1
      for key, value in target.items():
        try:
          temp[key] = peluang(att[key][tuple_X[i]],value+len(df[i].unique()))
        except Exception:
          temp[key] = peluang(1,value+len(df[i].unique()))
      prob[i] = temp

  #menghitung probe bersyarat untuk tiap kelas
  prob2 = {}
  for tar in target.keys():
    kali1 = 1
    for p in prob:
      kali1 *= prob[p][tar]
    prob2[tar] = kali1
  #Menghitung prob posterior bahwa tuple berada pada kelas C 
  bayes = {}
  for tar in target.keys():
    p = prob2[tar]*peluang_target[tar]
    bayes[tar] = -(1)/(np.log(p))
  for i,j in bayes.items():
    if j == max(bayes.values()):
      return i  
```

```
bayesian_learning(df,prediksi,attribut,'layak')
```

```
bayesian_learning(df,{
    'baterai':'Lemah',
    'kamera':'Tinggi',
    'harga':'Mahal'
}, attribut,'layak')
```
