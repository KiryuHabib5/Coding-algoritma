```
df_ya = df[df.layak == 'YA']
df_tidak = df[df.layak == 'Tidak']
```

```
def peluang(sample, banyak_data):
  return sample/banyak_data
```

```
prediksi = {'baterai':28,
            'kamera':4,
            'harga':2}
attribute = df.columns[1:-1]
```

```
target = df.groupby('layak')['handphone'].count()
```

```
def peluang_gaussian(X,std,mean):
  atas = np.e**(-(X-mean)**2/(2*std**2))
  bawah = std*np.sqrt(2*np.pi)
  return atas/bawah
```

```
kali = 1
for i in attribute:
  kali *= peluang_gaussian(prediksi[i],df_ya.std()[i],df_ya.mean()[i])
  print(peluang_gaussian(prediksi[i],df_ya.std()[i],df_ya.mean()[i]))
kali
```

```
def bayesian_learning(df,X,attribute,target):
  sample_target = df.groupby(target)['handphone'].count()
  
  #memisah berdasarkan kelas target (C)
  subtable = {}
  for key, value in sample_target.items():
    subtable[key] = df[df[target]==key]

  #Menghitung prob bersyarat (prob2)
  prob2 = {}
  try:
    for key, value in subtable.items():
      kali1 = 1
      #print(key," \n",value)
      for att in attribute:
        #print(att,key)
        #print("std : ",value.std()[att],"mean : ",value.mean()[att])
        kali1 *= peluang_gaussian(X[att],value.std()[att],value.mean()[att])
      p = peluang(sample_target[key],sample_target.sum())*kali1
      prob2[key] = -1/(np.log(p))
    print(prob2)
  except Exception:
    print(att)
    print(key,"\n",value)
  for i,j in prob2.items():
    if j == max(prob2.values()):
      return i  
```

```
atribut= df.columns[1:-1]
bayesian_learning(df,prediksi,atribut,'layak')
```
