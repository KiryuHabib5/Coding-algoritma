```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('/content/drive/My Drive/FolderBaru/Regresi/RM1_M1_S01_DS1_Linear Regression.csv')
df.shape
```

```
df = df.astype({'CLIENTNUM;Annual_Income;Age;Average_Monthly_Revenue':'str'})
```

```
df.dtypes
```

```
df[['CLIENTNUM','Annual_Income','Age','Average_Monthly_Revenue']] = df['CLIENTNUM;Annual_Income;Age;Average_Monthly_Revenue'].str.split(";",expand=True)
df.drop(columns='CLIENTNUM;Annual_Income;Age;Average_Monthly_Revenue',inplace=True)
```

```
df = df.apply(pd.to_numeric, errors='raise')
```

```
df.dtypes
```

```
df
```

```
df.corr()['Average_Monthly_Revenue']
```

```
y = df['Average_Monthly_Revenue'].values
x = df['Annual_Income'].values
```

```
x = x.reshape(-1,1)
x
```

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state = 42)
```

```
from sklearn.linear_model import LinearRegression
```

```
x_train.shape
```

```
model = LinearRegression()
model.fit(x_train, y_train)
```

```
model.score(x_test,y_test)
```

```
model.predict(np.array([2.2]).reshape(-1,1))
```

```
model.intercept_
```

```
model.predict(np.array([71500]).reshape(-1,1))
```

```
from sklearn.metrics import r2_score
```

```
r2_score(y_test, model.predict(x_test))
```
