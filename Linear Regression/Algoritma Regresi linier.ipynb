```
def gradien(x,y,x_mean, y_mean):
  atas = 0
  bawah = 0
  for i in range(0,len(x)):
    atas += x[i]*(y[i]-y_mean)

  for j in range(0,len(x)):
    bawah += x[j]*(x[j]-x_mean)
  return atas/bawah  
```

```
w = gradien(kamera,harga,np.mean(kamera),np.mean(harga))
```

```
a = np.mean(harga) - w*np.mean(kamera)
```
