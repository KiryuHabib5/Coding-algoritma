```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('/content/drive/My Drive/FolderBaru/Regresi/RM1_M2_S01_DS1_Multiple_Linear_Regression.csv')
df.head()
```

```
df[['ArrivalDelay','AirTime','DepartureDelay','Distance','TaxiIn','TaxiOut','MidOfWeekFlag']] = df['ArrivalDelay;AirTime;DepartureDelay;Distance;TaxiIn;TaxiOut;MidOfWeekFlag'].str.split(";", expand=True)
df.drop(columns='ArrivalDelay;AirTime;DepartureDelay;Distance;TaxiIn;TaxiOut;MidOfWeekFlag',inplace=True)
df.head()
```

```
df = df.apply(pd.to_numeric, errors='raise')
```

```
df.describe()
```

```
df.corr()['ArrivalDelay'].to_frame()
```

```
y = df['ArrivalDelay']
x = df[df.columns[1:]].values
```

```
y.shape
```

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.8, random_state=42)
```

```
from sklearn.linear_model import LinearRegression
```

```
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```
print('Coefficient : ' + str(lm.coef_))
print('Intercept : ' + str(lm.intercept_))
```

```
y_pred = lm.predict(x_test)
```
