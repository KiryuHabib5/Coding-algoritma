```
import numpy as np
import pandas as pd
```

```
df = pd.read_excel('/content/drive/My Drive/FolderBaru/Latihan Regresi.xlsx')
```

```
df.dropna(inplace=True) #Menghapus Missing values
```

```
df
```

```
x = np.array([(df['Baterai']),(df['Kamera'])])
x.shape
```

```
np.ones([1,6])
```

```
x
```

```
x_baru = np.vstack((np.ones([1,6]),x))
x_baru = x_baru.T
x_baru
```

```
y = df['Harga']
y_baru = np.array(y).reshape(-1,1)
y_baru
```

```
w = np.linalg.inv(x_baru.T@x_baru)@x_baru.T@y_baru
w
```

```
e = w[0]
w_baru = np.array(w[1:]).reshape(1,2)
print(e)
print(w_baru)
x_test = np.array([14,9])
print(x_test)
```

```
prediksi = w_baru@x_test + e
prediksi
```
