```
def pisah_rekomendasi(data):
  return [
          data[data['Layak Rekomendasi']=='YA'],
          data[data['Layak Rekomendasi']=='Tidak']
          ]
```

```
def P(sample,keseluruhan):
  return sample/keseluruhan
def entropy(Data):
  total = 0
  sigma = 0
  for i in range(len(Data)):
    #print(Data[i],"index ke-",i,"\n\n\n")
    #print(len(Data[i]))
    total += len(Data[i])
  for i in Data:
    E = -P(len(i),total)*np.log2(P(len(i),total))
    sigma +=E 
  return sigma
```

```
rekomendasi = pisah_rekomendasi(df)
ER = entropy(rekomendasi)
```

```
baterai = list(set(df['Baterai']))
```

```
cukup = df[df['Baterai']==baterai[0]]
Lemah = df[df['Baterai']==baterai[1]]
Kuat = df[df['Baterai']==baterai[2]]
```

```
E_baterai_kuat = entropy(pisah_rekomendasi(Kuat))
E_baterai_cukup = entropy(pisah_rekomendasi(cukup))
E_baterai_Lemah = entropy(pisah_rekomendasi(Lemah))
Entropy_Baterai = [E_baterai_kuat,E_baterai_cukup,E_baterai_Lemah]
Entropy_Baterai
```

```
#menghitung informasi gain bari baterai
ER - ((len(Kuat)/len(df))*E_baterai_kuat + (len(cukup)/len(df))*E_baterai_cukup + (len(Lemah)/len(df))*E_baterai_Lemah)
```
