```
import pandas as pd
import numpy as np
```

```
df = pd.read_excel('/content/drive/My Drive/FolderBaru/Decision Tree/Tabel Untuk Latihan Menghitung Entropy.xlsx')
```

```
df
```

```
df.drop(columns='Handphone',inplace=True)
```

```
baterai = pd.api.types.CategoricalDtype(['Kuat','Cukup','Lemah'],ordered=True) #Membuat tipe kategorical
kamera = pd.api.types.CategoricalDtype(['Tinggi','Sedang','Rendah'],ordered=True)
harga = pd.api.types.CategoricalDtype(['Sangat Mahal','Mahal','Murah','Sangat Murah'],ordered=True)
df['Baterai'] = df['Baterai'].astype(baterai)
df['Kamera'] = df['Kamera'].astype(kamera)
df['Harga'] = df['Harga'].astype(harga)
```

```
x = df[['Baterai','Kamera','Harga']]
y = df['Layak Rekomendasi']
```

```
x
```

```
import sklearn.tree 
```

```
model = sklearn.tree.DecisionTreeClassifier(criterion='gini')
```

```
model.fit(x,y)
```

```
model.predict([[2,2,1]])
```
