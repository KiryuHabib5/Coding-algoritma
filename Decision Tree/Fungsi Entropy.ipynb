```
def pisah_predict(data,kolom,key):
  bantu = []
  for i in key:
    bantu.append(data[data[kolom]==i])
  return bantu
```

```
target = pisah_predict(df,'Layak Rekomendasi',predict_list)
target
```

```
def Peluang(sample,keseluruhan):
  return sample/keseluruhan
def entropy(Data):
  total = 0
  sigma = 0
  for i in range(len(Data)):
    #print(Data[i],"index ke-",i,"\n\n\n")
    #print(len(Data[i]))
    total += len(Data[i])
  for i in Data:
    if len(i)==0:
      continue
    E = -Peluang(len(i),total)*np.log2(Peluang(len(i),total))
    sigma +=E 
  return sigma
```

```
Entropy_Target = entropy(target)
```

```
def gain(DataFrame,Target,kolom_values,key_values,kolom_predict,predict_key):
  Entropy_Target = entropy(Target)
  #print(Entropy_Target)
  bantu = pisah_predict(DataFrame,kolom_values,key_values)
  #print(bantu)
  #print("============================================================")
  #print(bantu[0])
  #print("========================================================")
  #print(bantu[1])
  #print("========================================================")
  #print(bantu[2])
  #print("========================================================")
  data_values = []
  for i in range(len(bantu)):
    data_values.append(pisah_predict(bantu[i],kolom_predict,predict_key))
  #print(data_values)
  #print("============================================================")
  
  sigma = 0
  for i in data_values:
    total = 0
    for j in range(len(i)):
      total+=len(i[j])
    sigma += (total/len(DataFrame))*entropy(i)
  return Entropy_Target-sigma
```
