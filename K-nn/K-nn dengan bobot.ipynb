```
def knn(df,attribute,target,input,k):
  bantu = 0
  for i in attribute:
    try:
      bobot = input[i]['bobot']
    except:
      bobot = 1
    try:
      bantu += bobot * (df[i]-input[i]['nilai'])**2
    except:
      bantu += bobot * (df[i]-input[i])**2
    df['Euclidean_distance'] = bantu**0.5
  df = df.sort_values(by='Euclidean_distance').reset_index(drop=True)
  df['masuk_kategori'] = None
  for i in range(len(df)):
    if i < k:
      df['masuk_kategori'][i] = True
    else:
      df['masuk_kategori'][i] = False
  prediksi = df.groupby(['masuk_kategori',target])[target].count()
  return prediksi[True]
```
